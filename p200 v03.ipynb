```python
import matplotlib.pyplot as plt
import numpy as np
import types # used to re-define acceleration / force

from mpl_toolkits.mplot3d.axes3d import Axes3D

%matplotlib inline
```

```python
class TimeSeries:
    def __init__(self, pts={}):
        self.pts = pts
            
    def update(self, pts={}):
        self.pts.update(pts)
 
    def append(self, time, value):
        self.pts[time] = value
 
    @property
    def timeline(self):
        return sorted(self.pts.keys())
    
    @property
    def values(self):
        return [self.pts[k] for k in self.timeline]
        
    def _get_time_intervals(self):
        t = self.timeline
        time_intervals = []
        for i in range(len(self.timeline) - 1):
            time_intervals.append((t[i], t[i + 1]))
        return time_intervals
    
    def _get_interval(self, time):
        for (t_min, t_max) in self._get_time_intervals():
            if time < t_min:
                return (None, t_min)
            if time < t_max:
                return (t_min, t_max)
        return (t_max, None)
    
    def roc(self, time):
        t_min, t_max = self._get_interval(time)
        if t_min is not None and t_max is not None:
            dp = self.pts[t_max] - self.pts[t_min]
            dt = t_max - t_min
            return dp / dt
        
    def __call__(self, time):
        t_min, t_max = self._get_interval(time)
        if t_min is not None and t_max is not None:
            return self.pts[t_min] + self.roc(time) * (time - t_min)
        else:
            return t_min, t_max

    def __getitem__(self, given):
        if isinstance(given, slice):
            timeline = self.timeline[given.start:given.stop:given.step]
            return [(t, self.pts[t]) for t in timeline]
        else:
            return self.pts[self.timeline[given]]
            
#     def __repr__(self):
#         return 
    
#     def plot(self, **kwargs):
#         plt.plot(self.timeline, self.values, **kwargs)
        
    def plot2d(self, **kwargs):
        n = len(self.values)
        lw = 1
        color = next(plt.axes()._get_lines.prop_cycler)['color']
        plt.plot(*[*zip(*self.values)][:2], **kwargs, c=color, lw=lw, 
                 markevery=[n-1], marker='o', mfc='w', mec=color, mew=lw)
        plt.axes().set_aspect('equal')
        
    def rate_of_change(self):
        roc = TimeSeries()
        for t_min, t_max in self._get_time_intervals():
            t_avg = ( t_max + t_min ) / 2
            roc.append(t_avg, self.roc(t_avg))
        return roc
            
#         return TimeSeries([(e[0], self.roc_at(e[0])) for e in self.events[1:]])
    
```
